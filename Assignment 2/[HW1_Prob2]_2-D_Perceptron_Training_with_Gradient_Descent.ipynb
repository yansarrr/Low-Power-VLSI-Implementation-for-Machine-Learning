```python
### Training with manually updating W with "Backward" ###
##Finished HW

import torch
import numpy as np
#from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim




data = [(1.0,2.1,3.0), (2.0, 3.5, 6.0), (3.0, 3.0, 9.0), (4.0, 2.1, 12.0), (5.0, 7.2, 15.0), (6.0, 10.1, 18.0)]


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 1, bias = False)
    def forward(self, x):
        x = self.fc1(x)
        return x

net = Net()

print(net)
print(list(net.parameters()))

#input = torch.randn(1)
#out = net(input)

#def criterion(out, label):
#    return (label - out)**2
criterion = nn.MSELoss()


optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
#optimizer = optim.Adam(net.parameters(), lr=0.005)

#----------------

batch_size = 3
num_epochs = 20
num_samples = len(data) #6

for epoch in range(num_epochs):
    np.random.shuffle(data)

    for i in range(0, num_samples, batch_size):
        mini_batch = data[i:i + batch_size]
        inputs = torch.FloatTensor([i[:2] for i in mini_batch])
        targets = torch.FloatTensor([i[-1] for i in mini_batch]).view(-1, 1)
        optimizer.zero_grad()
        outputs = net(inputs)
#         print("mini_batch", mini_batch)
#         print("inputs", inputs)
#         print("targets", targets)
#         print("outputs", outputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
    print("Epoch {} - loss: {}".format(epoch, loss))

for i in range(0, num_samples, batch_size):
        mini_batch = data[i:i + batch_size]
        inputs = torch.FloatTensor([i[:2] for i in mini_batch])
        targets = torch.FloatTensor([i[-1] for i in mini_batch]).view(-1, 1)
        optimizer.zero_grad()
        out = net(inputs)
        print("when x = {}, y = {}".format(inputs, out))

    
```
